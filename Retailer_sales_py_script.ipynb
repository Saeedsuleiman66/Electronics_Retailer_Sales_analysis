```python
customers = pd.read_csv('Customers.xlsx', sep=',', encoding='latin1')
products = pd.read_csv('Products.xlsx', sep=',', encoding='latin1')
sales = pd.read_csv('Sales.xlsx', sep=',', encoding='latin1')
exchange_rates = pd.read_csv('Exchange_Rates.xlsx', sep=',', encoding='latin1')
```

```python
for df in [customers, products, sales, exchange_rates]:df.columns=df.columns.str.strip()
```

```python
print(products.head(10))
```

```python
sales_enriched=pd.merge(sales,products,on='ProductKey',how='left')
```

```python
products.head(10)
```

```python
exchange_rates.head(10)
```

```python
sales_enriched.head(10)
```

```python
final_data=pd.merge(sales_enriched,exchange_rates,left_on=['Order Date','Currency Code'],right_on=['Date','Currency'],how='left')
```

```python
final_data.head(10)
```

```python
final_data = pd.merge(final_data, customers, on='CustomerKey', how='left')
```

```python
final_data.columns = [col.replace(' ', '_') for col in final_data.columns]
```

```python
final_data.head(10)
```

```python
cols_to_fix = ['Quantity', 'Unit_Price_USD', 'Unit_Cost_USD', 'Exchange']
for col in cols_to_fix:
    final_data[col] = pd.to_numeric(final_data[col].astype(str).str.replace(r'[^\d.]', '', regex=True), errors='coerce')
final_data[cols_to_fix] = final_data[cols_to_fix].fillna(0)
```

```python
final_data['Total_Revenue_USD'] = final_data['Quantity'] * final_data['Unit_Price_USD'] * final_data['Exchange']
final_data['Total_Cost_USD'] = final_data['Quantity'] * final_data['Unit_Cost_USD'] * final_data['Exchange']
final_data['Profit_USD'] = final_data['Total_Revenue_USD'] - final_data['Total_Cost_USD']
```

```python
final_data[['Total_Revenue_USD','Total_Cost_USD','Profit_USD']].head(10)
```

```python
final_data.duplicated().sum()
```

```python
final_data.isnull().sum()
```

```python
final_data['Delivery_Date']=final_data['Delivery_Date'].fillna('Not Shipped Yet')
final_data['State_Code'] = final_data['State_Code'].fillna('N/A')
```

```python
final_data.isnull().sum()
```

```python
final_data.to_csv('Global_Sales_Clean_Final.csv', index=False)
```
